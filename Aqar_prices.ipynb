```python
import pandas as pd

df = pd.read_csv("https://github.com/AFAskar/aqar-scraper/raw/refs/heads/main/aqar_fm_listings.csv")

```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
y = df[['price']]

X = df[['area_sqm','num_bedrooms', 'num_rooms', 'num_bathrooms', 'city', 'district']].copy()
```

```python
X.isnull().sum()
```

```python
# 1. Fill Numerical NaNs with Median
numerical_cols = ['area_sqm', 'num_bedrooms', 'num_rooms', 'num_bathrooms']
for col in numerical_cols:
    X[col] = X[col].fillna(X[col].median())

# 2. Fill Categorical NaNs with a placeholder
categorical_cols = ['city', 'district']
for col in categorical_cols:
    X[col] = X[col].fillna('Unknown')

# 3. One-Hot Encode the text columns
X_final = pd.get_dummies(X, columns=['city', 'district'])

# Check if any NaNs remain
print(X_final.isnull().sum().sum())
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Calculate MSE first
mse = mean_squared_error(y_test, y_pred)

# Take the square root to get RMSE
rmse = np.sqrt(mse)

print(f"On average, our model is off by: {rmse:,.2f} SAR")
```

```python
# Filtering to keep only sales
df_sale = df[df['is_sale'] == True]
df_sale.shape
```

```python
y = df_sale[['price']]

X = df_sale[['area_sqm','num_bedrooms', 'num_rooms', 'num_bathrooms', 'city', 'district']].copy()
```

```python
X.isnull().sum()
```

```python
# 1. Fill Numerical NaNs with Median for the sales data
numerical_cols = ['area_sqm', 'num_bedrooms', 'num_rooms', 'num_bathrooms']
for col in numerical_cols:
    X[col] = X[col].fillna(X[col].median())

# 2. Fill Categorical NaNs
categorical_cols = ['city', 'district']
for col in categorical_cols:
    X[col] = X[col].fillna('Unknown')

# 3. Create the final version with One-Hot Encoding
X_final = pd.get_dummies(X, columns=['city', 'district'])

# Verify the result
print(X_final.isnull().sum().sum())
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate MSE
mse_new = mean_squared_error(y_test, y_pred)

# Calculate RMSE
rmse_new = np.sqrt(mse_new)

print(f"New RMSE: {rmse_new:,.2f} SAR")
```

### Trying a different model

```python
from sklearn.ensemble import RandomForestRegressor



rf_model = RandomForestRegressor()



rf_model.fit(X_train, y_train.values.ravel())
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate RMSE for the Random Forest
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))

print(f"Random Forest RMSE: {rmse_rf:,.2f} SAR")
```

```python
# Create a comparison table for the first 5 test samples
comparison_df = y_test.head(5).copy()
comparison_df['Predicted_Price'] = y_pred_rf[:5]
comparison_df['Difference'] = comparison_df['price'] - comparison_df['Predicted_Price']

print(comparison_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_rf, alpha=0.5, color='royalblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Price (SAR)')
plt.ylabel('Predicted Price (SAR)')
plt.title('Actual vs Predicted House Prices (Random Forest)')
plt.show()
```
